## Natural Language to Simplified App Representation (SAR)

**Please describe an app in natural language using the following components:**

`'textbox', 'button', 'text2speech', 'ball', 'accelerometer', 'video_player', 'switch', 'player', 'label', 'timepicker', 'passwordtextbox'`

Some sample *event*, *action*, and *value* are below,

**Events**: 
- button click, 
- switch flip, 
- accelerometer shaken, 
- ball flung

**Actions**: 
- Start, stop, video/audio
- Speak a given text or a textbox
- Bounce ball, set speed/color of the ball, etc.
- Set label

**Value**:
- textbox text
- any literal, integer
- time picker time

**Assets**: Text2App/Media/



Example description:

***Create a mobile application with a textbox, a button, and a text2speech. When the button is pressed, say the text in the text box.***

Available Assets in Media directory

Any new file can be kept in this directory and used for app creation

```python
!tree Media
```

```python
NL = 'Create an app with a textbox, a button named "Speak", and a text2speech. When the button is clicked, speak the text in the text box.'
```

```python
t2a = Text2App(NL, nlu='roberta') # NLU options: 'pointernet' (default), 'roberta', 'codebert'
```

**Processed Natural Language**

```python
t2a.NL
```

```python
t2a.SAR
```

```python
print(t2a.prettyprint_sar())
```

```python
t2a.literal_dict
```

## Compile SAR to Application source

```python
app_name = "Speak_it"
```

```python
sar_to_aia(t2a, project_name=app_name)
```
